```python
import random

# Funktion zum Lesen der Paketdaten aus einer Datei
def read_package_data(file_name):
    packages = []
    with open(file_name, "r") as file:
        for line in file:
            parts = line.strip().split(";")
            packages.append({
                "id": int(parts[0].split()[1]),
                "value": int(parts[1].replace("EUR", "").strip()),
                "weight": int(parts[2].replace("kg", "").strip())
            })
    return packages

# Paketdaten aus der Datei lesen
packages = read_package_data("data/Pakete.txt")

max_weight = 26630
population_size = 50
generations = 100
mutation_rate = 0.1

# Initialisiere die Population
def initialize_population(size):
    return [[random.choice([0, 1]) for _ in range(len(packages))] for _ in range(size)]

# Fitnessfunktion
def calculate_fitness(individual):
    total_weight = sum(p["weight"] * individual[i] for i, p in enumerate(packages))
    total_value = sum(p["value"] * individual[i] for i, p in enumerate(packages))
    if total_weight > max_weight:
        return 0
    return total_value

# Selektionsfunktion
def selection(population):
    weighted_population = [(individual, calculate_fitness(individual)) for individual in population]
    weighted_population.sort(key=lambda x: x[1], reverse=True)
    return [individual for individual, fitness in weighted_population[:population_size//2]]

# Kreuzung
def crossover(parent1, parent2):
    crossover_point = random.randint(0, len(parent1) - 1)
    return parent1[:crossover_point] + parent2[crossover_point:]

# Mutation
def mutation(individual):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] = 1 if individual[i] == 0 else 0
    return individual

# Hauptfunktion des genetischen Algorithmus
def genetic_algorithm():
    population = initialize_population(population_size)
    for generation in range(generations):
        population = selection(population)
        new_population = []
        while len(new_population) < population_size:
            parent1, parent2 = random.sample(population, 2)
            child = crossover(parent1, parent2)
            child = mutation(child)
            new_population.append(child)
        population = new_population
    best_solution = max(population, key=calculate_fitness)
    return best_solution

# Ausführung des genetischen Algorithmus und Ausgabe der besten Lösung
best_solution = genetic_algorithm()
best_packages = [packages[i] for i in range(len(best_solution)) if best_solution[i] == 1]

# Beste Pakete in eine Datei schreiben und im Terminal ausgeben
with open("Optimal_Packages.txt", "w") as file:
    for package in best_packages:
        output = f"Package {package['id']};{package['value']}EUR;{package['weight']}kg"
        file.write(output + "\n")
        print(output)

# Gesamtwert der besten Lösung ausgeben
total_value = sum(p["value"] for p in best_packages)
print(f"Total value: {total_value} EUR")

# Gesamtgewicht der besten Lösung ausgeben
total_weight = sum(p["weight"] for p in best_packages)
print(f"Total weight: {total_weight} kg")
```
